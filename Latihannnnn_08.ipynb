Praktikum ini merupakan rangkaian lengkap proses Pembersihan Data (Data Cleaning) yang bertujuan untuk mengubah data mentah yang semula tidak konsisten menjadi data yang valid, rapi, dan lengkap. Fokus pembersihan dilakukan pada dua kolom utama:

1. Kolom jurusan (Data Kategorikal)

Pada kolom ini dilakukan proses normalisasi penulisan, yaitu penyamaan format teks agar semua penulisan jurusan menjadi seragam. Perbedaan huruf besar–kecil seperti “Teknik” dan “teknik” yang sebelumnya dianggap kategori berbeda telah digabungkan menjadi satu kategori yang konsisten. Ini memastikan hasil analisis kategorikal tidak bias atau terpecah.

2. Kolom umur (Data Numerik)

Pembersihan pada kolom ini dilakukan lebih mendalam, karena ditemukan cukup banyak ketidakkonsistenan, seperti teks, angka tidak logis, serta nilai kosong.

Langkah-langkah yang dilakukan meliputi:

Koreksi Nilai: Mengubah nilai teks seperti “twenty” dan “21years” menjadi angka yang benar.

Mengatasi Nilai Tidak Valid: Nilai yang tidak logis seperti -5, 150, atau teks “NaN_string” kemudian diganti dengan np.nan agar bisa diproses.

Imputasi Rata-rata: Seluruh nilai hilang dan nilai tidak valid diganti dengan rata-rata umur, yang dibulatkan menjadi 22 untuk konsistensi data.

Verifikasi Akhir: Melalui isna().sum(), dipastikan bahwa tidak ada lagi nilai yang kosong dan semua nilai pada kolom umur telah valid serta seragam.
